# This Program uses pyenchant as a spellchecker to autocorrect inputed strings

```python
!pip install pyenchant
import enchant
import pandas as pd
import numpy as np
import csv
#Pyenchant comes standard in anaconda on ubuntu but doesn't run on windows. All other packages are necessary
```

```python
dictionary = pd.read_csv('/home/michael/Downloads/felsndc.csv')
#Import cvs file containing list used as dic
```

```python
dictionary.head() #Here is the imported csv
```

```python
drug_name = dictionary['drug_name'].values 
#Now we take the csv variable and pull out the variable of interest
```

```python
np.savetxt('drug_name.txt', drug_name, fmt ='%s') 
#Save the variable into a .txt file
```

```python
pwl = enchant.request_pwl_dict('drug_name.txt') 
#Import the .txt into enchant as a custom made dictionary
```

```python
pwl.check('ANESTHETIC') 
#The spellcheck is able to recognize exact matches
```

```python
pwl.check('Anesthetic') 
#Spellcheck is case sensitive. Dictionary and Variables being checked will need to be put into upercase
```

```python
pwl.suggest('Anesthetic')
#This functions outputs a suggestion 
```

```python
code_string = (pd.Series.to_string(dictionary.Code))
#For the next section of code we need to turn our variable into a string
#This line takes the code variable of the dictionary data set 
#and converts it into a string function
```

```python
from enchant.checker import SpellChecker
chkr = SpellChecker(pwl)
chkr.set_text(code_string)
for err in chkr:
    print (err.word) #Option to print the wrong word
    sug = err.suggest()[0]
    err.replace(sug)
corrected = chkr.get_text()
print (corrected)
#This is not my module, credit goes to Roy Holzem on stackoverflow
#URL: https://stackoverflow.com/questions/31026394/how-to-correct-text-and-return-the-corrected-text-automatically-with-pyenchant
#Corrected will output as a data string but input must be a single string
```
